Punto 1

```python
import numpy as np

num_candidatos = 30
total_votos = 5000

# Inicializar contadores de votos en cero
votos = np.zeros(num_candidatos, dtype=int)

# Asignar 5000 votos aleatoriamente a los 30 candidatos
for _ in range(total_votos):
    idx = np.random.randint(0, num_candidatos)  # Elegir un candidato aleatorio
    votos[idx] += 1  # Sumar un voto

# Ordenar candidatos de mayor a menor segun la cantidad de votos
candidatos = np.arange(1, num_candidatos + 1)  # Numeros de los candidatos (1 a 30)
orden = np.argsort(votos)[::-1]  # Índices ordenados de mayor a menor votos


# Verificar la suma total
print("Listado de candidatos de mayor a menor número de votos:")
for i in range(num_candidatos):
    print(f"Candidato {candidatos[orden[i]]}: {votos[orden[i]]} votos")
```

```python
import numpy as np

# Definir número de estudiantes
num_estudiantes = 6500

# Generar datos simulados
np.random.seed(42)  # Semilla para reproducibilidad

# Años de ingreso entre 1980 y 2025
años_ingreso = np.random.randint(1980, 2026, num_estudiantes)

# Convertir años al formato deseado (ej. 1998 → 198, 2021 → 221)
primer_digito = np.where(años_ingreso < 2000, 1, 2)  # 1 para 1900s, 2 para 2000s
ultimos_dos_digitos = años_ingreso % 100  # Extraer los últimos dos dígitos

# Construir los 3 primeros dígitos del código
codigo_año = primer_digito * 100 + ultimos_dos_digitos

# Últimos 4 dígitos del código (aleatorios entre 1000 y 9999)
numeros_estudiante = np.random.randint(1000, 10000, num_estudiantes)

# Construir el código de 7 dígitos
codigos = codigo_año * 10**4 + numeros_estudiante

# Generar otros datos
nombres = np.array([f"Estudiante {i}" for i in range(num_estudiantes)])  # Nombres ficticios
carreras = np.random.randint(1, 10, num_estudiantes)  # Carreras representadas por un código (1-9)
promedios = np.round(np.random.uniform(2.0, 5.0, num_estudiantes), 2)  # Promedio entre 2.0 y 5.0

# 1. Filtrar estudiantes por carrera y promedio >= 4.0
codigo_carrera = int(input("Ingrese el código de la carrera a listar: "))
filtro_carrera = (carreras == codigo_carrera) & (promedios >= 4.0)

estudiantes_filtrados = np.column_stack((codigos[filtro_carrera], nombres[filtro_carrera]))

print("\nEstudiantes de la carrera con promedio >= 4.0:")
for codigo, nombre in estudiantes_filtrados:
    print(f"{codigo} - {nombre}")

print(f"\nTotal de estudiantes con promedio >= 4.0 en la carrera {codigo_carrera}: {len(estudiantes_filtrados)}")

# 2. Filtrar estudiantes que ingresaron antes de 1990 y están condicionales
filtro_condicionales = (años_ingreso < 1990) & (promedios < 3.2) & (promedios >= 2.7)

estudiantes_condicionales = np.column_stack((codigos[filtro_condicionales], nombres[filtro_condicionales]))

print("\nEstudiantes que ingresaron antes de 1990 y están condicionales:")
for codigo, nombre in estudiantes_condicionales:
    print(f"{codigo} - {nombre}")

print(f"\nTotal de estudiantes condicionales que ingresaron antes de 1990: {len(estudiantes_condicionales)}")
```
